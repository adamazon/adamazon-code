#### NearestNeighbors
We can try to use a NearestNeighbors method to find similar product. However, there is only the price column that is a number.  

##### Categories
We transform the categories to a one hot encoding format.  
An Amazon categories is represented by a list of categories, sub categories, subsub categories ...

For example `['Electronics', 'Television & Video', 'Blu-ray Players & Recorders']` or `['Video Games', 'PlayStation 4', 'Consoles']`. The leftest category is the most specific.

We can encode a list `['Electronics', 'Television & Video', 'Blu-ray Players & Recorders']` as a category, but we can also consider the subcategory as a category to. So we will have `'Electronics'`, `'Television & Video'`and `'Blu-ray Players & Recorders'`. We will use this representation as it allows more combinations and makes more easier the neighbor search. 

For example `CategoryA -> SubCategoryZ -> SubsubCategoryA` is very similar to `CategoryA -> SubCategoryY -> SubsubCategoryA`. Only the `Subcategory` makes it different.

```python
flat_list_categories = [item for sublist in video_games_meta.categories.values for item in sublist] # Flatten the list of categories
flat_list_categories = [item for sublist in flat_list_categories for item in sublist] # Flatten the categories, subcategories etc..
flat_list_categories = list(set(flat_list_categories)) # Remove duplicates
flat_list_categories[:10]
```

```python
print("We get", len(flat_list_categories), "number of categories")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

# Encode category string to number
label_encoder.fit(flat_list_categories)
categories = label_encoder.transform(flat_list_categories)

# Encode to one hot
on_hot_encoder = OneHotEncoder()
on_hot_encoder.fit([[cat] for cat in categories])
```

```python
# The one hot encoding for the 223 category
on_hot = np.asarray(on_hot_encoder.transform(223).todense()).reshape(-1)
on_hot
```

```python
def get_features_from_product_metadata(product):
    categories_labels = []
    for category_list in product.categories:
        categories_labels = np.concatenate([categories_labels, label_encoder.transform(category_list)])
    
    categories_one_hot = np.zeros(468) # 468 categories
    for cat_label in categories_labels:
        categories_one_hot = categories_one_hot + on_hot_encoder.transform(cat_label)
        
    features = list(np.asarray(categories_one_hot[0])[0]) # Categories features
    features.append(product.price) # Add price feature
    return features

# Create a list of products with onehot category and price features

video_games_meta_dropped_price_nan = video_games_meta[np.isfinite(video_games_meta['price'])]

X = []
X_asin = []
for i, product in video_games_meta_dropped_price_nan.iterrows():
    features = get_features_from_product_metadata(product)
    
    X.append(features)
    X_asin.append(product.asin)
    
    if i % 1000 == 0:
        print(i,"/",len(video_games_meta_dropped_price_nan))
```

We now can use these products representation to find the nearest neighbors for a given product

```python
# Return the indices of neighbors products
def get_nearest_neighbors_product(product, X, X_asin, n_neighbors=2):
    nbrs = NearestNeighbors(n_neighbors=n_neighbors, algorithm='ball_tree').fit(X)
    distances, indices = nbrs.kneighbors([get_features_from_product_metadata(product.squeeze())])
    
    asins = [X_asin[i] for i in indices[0]]
    return asins
```

```python
asins = get_nearest_neighbors_product(playstation, X, X_asin)
```

```python
video_games_meta[video_games_meta.asin == asins[1]]
```

The similar product obtained for the playstation console is a James bond game. We would expect to get a similar console. This is due to the playstation price of $28.12, which is probably a mistake.

We try it on a Xbox game :

```python
xbox_game = video_games_meta.iloc[0]
```

```python
asins = get_nearest_neighbors_product(xbox_game, X, X_asin)
video_games_meta[video_games_meta.asin.isin(asins)]
```

We get a similar product, which stay in the same category "Xbox360 games".

#### Get enough products to reach 10k reviews
We start from one xbox game and get 1000 nearest neighbors

```python
asins = get_nearest_neighbors_product(xbox_game, X, X_asin, 1000)
video_games_meta[video_games_meta.asin.isin(asins)].head()
```

Then we concatenate the reviews of the 1000 xbox video games

```python
xbox_games_reviews = video_games.iloc[video_games.index.get_level_values('asin').isin(asins)]
print("Reviews count:", len(xbox_games_reviews))
xbox_games_reviews.head()
```
